```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "/Users/armando_yaelm/desktop/schools_complete.csv"
student_data_to_load = "/Users/armando_yaelm/desktop/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name"])
```

```python
#average math scores
avg_math= student_data["math_score"].mean()
avg_math
```

```python
#%passing math
math_70 = student_data["math_score"] > 69
true_count_m = math_70.sum()
passing_math = (true_count_m / total_students) * 100
passing_math
```

```python
#% Passing Reading
reading_70 = student_data["reading_score"] > 69
true_count_r = reading_70.sum()
passing_reading = (true_count_r / total_students) * 100
passing_reading
```

```python
#overall passing rate
overall = (passing_reading + passing_math) /2
overall
```

```python

df=pd.DataFrame(performance_schools)
bins = [0,69,100]
group_names = [0,1]

df["aprobados_mate"] = pd.cut(df["math_score"], bins, labels=group_names)
grupo_mate = df[["aprobados_mate", "school_name"]].groupby(["school_name"])
aprobaron_mate=(grupo_mate.sum() /grupo_mate.count()) *100
aprobaron_mate

```

```python
df["aprobados_lectura"] = pd.cut(df["reading_score"], bins, labels=group_names)
grupo_lectura = df[["aprobados_lectura", "school_name"]].groupby(["school_name"])
aprobaron_lectura=(grupo_lectura.sum() /grupo_lectura.count() ) *100
#school_name = performance_comparison.index
aprobaron_lectura
```

```python

df=pd.DataFrame(performance_schools)
bins = [0,69,100]
group_names = [0,1]

df["aprobados_mate"] = pd.cut(df["math_score"], bins, labels=group_names)
df["aprobados_lectura"] = pd.cut(df["reading_score"], bins, labels=group_names)

df['% Overall Passing Rate'] = df.apply(lambda x: x['aprobados_mate'] + x['aprobados_lectura'], axis=1)
df

grupo_suma = df[["% Overall Passing Rate", "school_name"]].groupby(["school_name"])

totales = grupo_suma.sum() / (grupo_mate.sum() + grupo_lectura.sum())
totales = (grupo_suma.sum() / (grupo_suma.count()*2))*100

totales
```
